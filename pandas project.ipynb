```python
# Pandas is used for data processing and analysis
# Used to create 2-dimensional tabular data structure with labled axis
import pandas as pd                         # importing Pandas library under the alias "pd"
dataframe = pd.read_csv("Name/path of file")       # read csv file
dataframe = pd.read_excel("Name/path of file")     # read excel file
dataframe.to_csv("file name")                      # converts file to csv format
dataframe.shape                                    # gives shape, i.e number of rows and columns in dataframe
dataframe.info                                     # give information about dataframe
dataframe.value_counts("column name")              # gives the count of different values in column
dataframe.mean()                                   # gives mean of each column
dataframe.describe                                 # gives all statistical measures of dataframe
dataframe.drop(index=0,axis=0)                     # to remove row axis should be 0 and for column it should be 1
dataframe.drop(columns="column name",axis=1)       # remove column
```
